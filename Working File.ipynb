```python
import yfinance as yf

# List of tickers and industry
tickers = ["AAPL", "NVDA", "GOOGL", "AMZN", "META"]
industry = "Technology"

# Loop through each ticker
for ticker in tickers:
    dat = yf.Ticker(ticker)
    df1 = dat.history(period='1mo')

    # Reset index to make 'Date' a column
    df1 = df1.reset_index()

    # Add a new column with the ticker name
    df1["Ticker"] = ticker
    df1["Industry"] = industry

    # Remove unnecessary columns
    df1 = df1.drop(columns=["Dividends", "Stock Splits"])

    # Display the DataFrame
    print(df1)
```
