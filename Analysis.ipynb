```python
def plot_volume_distribution(df):
    sns.set(style="whitegrid")
    
    plt.figure(figsize=(10, 5))
    sns.histplot(df['Volume'], kde=True, bins=30, color='green')
    plt.title('Distribution of Trading Volume', fontsize=14)
    plt.xlabel('Volume', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.tight_layout()
    plt.show()

# Call the function
plot_volume_distribution(df)
```

```python
from scipy.stats import ttest_ind
from scipy.stats import pearsonr

price_high = df['High'].dropna()
price_low = df['Low'].dropna()

# Ensure both series have the same length for comparison
min_length = min(len(price_high), len(price_low))
price_high = price_high[:min_length]
price_low = price_low[:min_length]

t_stat, p_value = ttest_ind(price_high, price_low, equal_var=False)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```
